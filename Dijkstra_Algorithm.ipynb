```python
# Define the costs column in the table
infinity = float("inf")
costs = {}
costs["current_location"] = 0
costs["a"] = infinity
costs["b"] = infinity
costs["c"] = infinity
costs["d"] = infinity
costs["e"] = infinity
costs["f"] = infinity
costs["g"] = infinity
costs["h"] = infinity
costs["i"] = infinity
costs["j"] = infinity
costs["k"] = infinity
costs["l"] = infinity
costs["m"] = infinity
costs["n"] = infinity
costs["o"] = infinity
costs["p"] = infinity
costs["q"] = infinity
costs["r"] = infinity
costs["s"] = infinity
costs["t"] = infinity
costs["u"] = infinity
costs["v"] = infinity
costs["toilet_A"] = infinity
costs["toilet_B"] = infinity
costs["toilet_C"] = infinity
costs["toilet_D"] = infinity

# Define the parents column in the table
parents = {}
parents["a"] = None
parents["b"] = None
parents["c"] = None
parents["d"] = None
parents["e"] = None
parents["f"] = None
parents["g"] = None
parents["h"] = None
parents["i"] = None
parents["j"] = None
parents["k"] = None
parents["l"] = None
parents["m"] = None
parents["n"] = None
parents["o"] = None
parents["p"] = None
parents["q"] = None
parents["r"] = None
parents["s"] = None
parents["t"] = None
parents["u"] = None
parents["v"] = None
parents["toilet_A"] = None
parents["toilet_B"] = None
parents["toilet_C"] = None
parents["toilet_D"] = None

#Define the processed column in the table
processed = []
```

```python
# This function finds the node with lowest cost in the costs column.
def find_lowest_cost_node(costs):
    lowest_cost = float("inf")
    lowest_cost_node = None
    # Go through each node.
    for node in costs:
        cost = costs[node]
        # If it's the lowest cost so far and hasn't been processed yet...
        if cost < lowest_cost and node not in processed:
            # ... set it as the new lowest-cost node.
            lowest_cost = cost
            lowest_cost_node = node
    return lowest_cost_node
```

```python
# Find the lowest-cost node that you haven't processed yet.
node = find_lowest_cost_node(costs)
# If you've processed all the nodes, this while loop is done.
while node is not None:
    cost = costs[node]
    # Go through all the neighbors of this node.
    neighbors = graph[node]
    for n in neighbors.keys():
        new_cost = cost + neighbors[n]
        # If it's cheaper to get to this neighbor by going through this node...
        if costs[n] > new_cost:
            # ... update the cost for this node.
            costs[n] = new_cost
            # This node becomes the new parent for this neighbor.
            parents[n] = node
    # Mark the node as processed.
    processed.append(node)
    # Find the next node to process, and loop.
    node = find_lowest_cost_node(costs)
```

```python
# This block simulates FlushFinder asking the user to choose a toilet and giving the user the shortest path to it.
# The code will ask you to input the toilet location you want to go to. Please type the location according to the instructions.
# This can also give the shortest path to ANY node from the starting location.
# A recursion may occur if the code above is not run first. Breaks have been included to deal with this.

while True:
  try:
    chosen_location = input("Which toilet do you want to go? (Type either 'toilet_A', 'toilet_B', 'toilet_C' or 'toilet_D') ")
    shortest=[chosen_location]
    node=chosen_location

    while node != "current_location":
      parent=parents[node]
      shortest.append(parent)
      node = parent

    shortest.reverse()
    print("You have chosen "+chosen_location+".")
    print("The shortest path to "+ chosen_location +" of length " +str(costs[chosen_location]) +" and given by " +str(shortest)+ ".")
    print("Re-run this chunk to search for another toilet.")
    break

  except:
    print("Error. Please type in the toilet you want to go to exactly as the instruction. Re-run this chunk.")
    break
```
